```
y = int(input("薬草の数は？"))
d = int(input("毒消の数は？"))
w = float(input("割引率は？"))
#関数の定義
def calc(yn, dn, wa):
    return (yn * 8 + dn * 10) * wa

print("薬草が{0}個、毒消が{1}個で{2}Gです".format(y, d, calc(y, d, w)))
```
